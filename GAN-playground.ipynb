## Let's Break Some Generative Adversarial Networks!

You'll need to install [PyTorch](http://pytorch.org/), `numpy`, and `matplotlib`

First, we'll look at a simple implementation of a GAN from https://arxiv.org/abs/1406.2661 on MNIST, code based on https://github.com/yunjey/pytorch-tutorial/blob/master/tutorials/02-intermediate/generative_adversarial_network/main.py. 

The first time you run this code, set `DOWNLOAD` to `True`.

```python
import torch
import torchvision
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets 
from torchvision import transforms
from torchvision.utils import save_image
from torch.autograd import Variable
import matplotlib.pyplot as plt
import numpy as np

DOWNLOAD = False

# Added by Patrick for showing images inline
%matplotlib inline
def show(img, ax):
    if torch.cuda.is_available():
        img = img.cpu()
    npimg = img.numpy()
    ax.imshow(np.transpose(npimg, (1,2,0)), interpolation='nearest')
    
def to_var(x):
    if torch.cuda.is_available():
        x = x.cuda()
    return Variable(x)

def denorm(x):
    out = (x + 1) / 2
    return out.clamp(0, 1)

# Image processing 
transform = transforms.Compose([
                transforms.ToTensor(),
                transforms.Normalize(mean=(0.5, 0.5, 0.5), 
                                     std=(0.5, 0.5, 0.5))])
# MNIST dataset
mnist = datasets.MNIST(root='./data/',
                       train=True,
                       transform=transform,
                       download=DOWNLOAD)
# Data loader
data_loader = torch.utils.data.DataLoader(dataset=mnist,
                                          batch_size=100, 
                                          shuffle=True)
# Discriminator
D = nn.Sequential(
    nn.Linear(784, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 1),
    nn.Sigmoid())

# Generator 
G = nn.Sequential(
    nn.Linear(64, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 784),
    nn.Tanh())

if torch.cuda.is_available():
    D.cuda()
    G.cuda()

# Binary cross entropy loss and optimizer
criterion = nn.BCELoss()
d_optimizer = torch.optim.Adam(D.parameters(), lr=0.0003)
g_optimizer = torch.optim.Adam(G.parameters(), lr=0.0003)

# We're going to plot the D's error and accuracy
d_err = [[] for _ in range(3)]
d_acc = [[] for _ in range(3)]

# Start training
for epoch in range(150):
    for i, (images, _) in enumerate(data_loader):
        # Build mini-batch dataset
        batch_size = images.size(0)
        images = to_var(images.view(batch_size, -1))
        
        # Create the labels which are later used as input for the BCE loss
        real_labels = to_var(torch.ones(batch_size))
        fake_labels = to_var(torch.zeros(batch_size))

        #============= Train the discriminator =============#
        # Compute BCE_Loss using real images where BCE_Loss(x, y): - y * log(D(x)) - (1-y) * log(1 - D(x))
        # Second term of the loss is always zero since real_labels == 1
        outputs = D(images)
        d_loss_real = criterion(outputs, real_labels)
        real_score = outputs
        
        # Compute BCELoss using fake images
        # First term of the loss is always zero since fake_labels == 0
        z = to_var(torch.randn(batch_size, 64))
        fake_images = G(z)
        outputs = D(fake_images)
        d_loss_fake = criterion(outputs, fake_labels)
        fake_score = outputs
        
        # Backprop + Optimize
        d_loss = d_loss_real + d_loss_fake
        D.zero_grad()
        d_loss.backward()
        d_optimizer.step()
        
        if (epoch+1) == 25:     
            d_err[0].append(d_loss.data[0])
            d_acc[0].append(fake_score.data[0,0])
        elif (epoch+1) == 100:
            d_err[1].append(d_loss.data[0])
            d_acc[1].append(fake_score.data[0,0])
        elif (epoch+1) == 150:
            d_err[2].append(d_loss.data[0])
            d_acc[2].append(fake_score.data[0,0])
        #=============== Train the generator ===============#
        # Compute loss with fake images
        z = to_var(torch.randn(batch_size, 64))
        fake_images = G(z)
        outputs = D(fake_images)
        
        # We train G to maximize log(D(G(z)) instead of minimizing log(1-D(G(z)))
        # For the reason, see the last paragraph of section 3. https://arxiv.org/abs/1406.2661
        g_loss = criterion(outputs, real_labels)
        
        # Backprop + Optimize
        D.zero_grad()
        G.zero_grad()
        g_loss.backward()
        g_optimizer.step()
        
        if (i+1) % 300 == 0:
            print('Epoch [%d/%d], Step[%d/%d], d_loss: %.4f, '
                  'g_loss: %.4f, D(x): %.2f, D(G(z)): %.2f' 
                  %(epoch, 150, i+1, 600, d_loss.data[0], g_loss.data[0],
                    real_score.data.mean(), fake_score.data.mean()))
    
    # Show a few real images
    if (epoch+1) == 1:
        images = images.view(images.size(0), 1, 28, 28)
        # save_image(denorm(images.data), './data/real_images.png')
        fig, axs = plt.subplots(1,3)
        plt.suptitle('real images')
        for j in range(3):
            img = denorm(images[j].data)
            img = torch.cat([img, img, img], 0).float()
            show(img, axs[j])
    
    if (epoch+1) % 50 == 0:
        # Save sampled images
        fake_images = fake_images.view(fake_images.size(0), 1, 28, 28)
        #save_image(denorm(fake_images.data), './data/fake_images-%d.png' %(epoch+1))
        fig, axs = plt.subplots(1,3)
        plt.suptitle('fake images after epoch {}'.format(epoch+1))
        for j in range(3):
            img = denorm(fake_images[j].data)
            img = torch.cat([img, img, img], 0).float()
            show(img, axs[j])

fig, ax = plt.subplots()
plt.ylabel('BCE')
plt.xlabel('training iterations')
plt.title('Discriminators error')
plt.plot(d_err[0], '-', c='r', label='After 25 epoch')
plt.plot(d_err[1], '--', c='b', label='After 100 epochs')
plt.plot(d_err[2], ':', c='g', label='After 150 epochs')
plt.legend()

fig, ax = plt.subplots()
plt.ylabel('Accuracy')
plt.xlabel('training iterations')
plt.title('Discriminators accuracy')
plt.plot(d_acc[0], '-', c='r', label='After 25 epoch')
plt.plot(d_acc[1], '--', c='b', label='After 100 epochs')
plt.plot(d_acc[2], ':', c='g', label='After 150 epochs')
plt.legend()
```

## Case when generative and real data distributions are disjoint

Now, let's mess with $p_g$ and $p_{data}$. If we forcibly alter the images generated by $G$, then theoretically $p_g$ and $p_{data}$ will be disjoint and $D$ should converge rapidly.

```python
import torch.nn.functional as F

# We'll instantiate new models first
# Discriminator
D = nn.Sequential(
    nn.Linear(784, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 1),
    nn.Sigmoid())

# Generator 
G = nn.Sequential(
    nn.Linear(64, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 256),
    nn.LeakyReLU(0.2),
    nn.Linear(256, 784),
    nn.Tanh())

if torch.cuda.is_available():
    D.cuda()
    G.cuda()

# Binary cross entropy loss and optimizer
criterion = nn.BCELoss()
d_optimizer = torch.optim.Adam(D.parameters(), lr=0.0003)
g_optimizer = torch.optim.Adam(G.parameters(), lr=0.0003)

# We're going to plot the D's error and accuracy
d_err = [[] for _ in range(3)]
d_acc = [[] for _ in range(3)]

# Start training
for epoch in range(5):
    for i, (images, _) in enumerate(data_loader):
        # Build mini-batch dataset
        batch_size = images.size(0)
        images = to_var(images.view(batch_size, -1))
        
        # Create the labels which are later used as input for the BCE loss
        real_labels = to_var(torch.ones(batch_size))
        fake_labels = to_var(torch.zeros(batch_size))

        #============= Train the discriminator =============#
        # Compute BCE_Loss using real images where BCE_Loss(x, y): - y * log(D(x)) - (1-y) * log(1 - D(x))
        # Second term of the loss is always zero since real_labels == 1
        outputs = D(images)
        d_loss_real = criterion(outputs, real_labels)
        real_score = outputs
        
        # Compute BCELoss using fake images
        # First term of the loss is always zero since fake_labels == 0
        z = to_var(torch.randn(batch_size, 64))
        fake_images = G(z)
        ################
        ## Let's invert the fake_images here
        ################
        outputs = D(0 * fake_images)
        d_loss_fake = criterion(outputs, fake_labels)
        fake_score = outputs
        
        # Backprop + Optimize
        d_loss = d_loss_real + d_loss_fake
        D.zero_grad()
        d_loss.backward()
        d_optimizer.step()
        
        if (epoch+1) == 1:     
            d_err[0].append(d_loss.data[0])
            d_acc[0].append(fake_score.data[0,0])
        elif (epoch+1) == 3:
            d_err[1].append(d_loss.data[0])
            d_acc[1].append(fake_score.data[0,0])
        elif (epoch+1)== 5:
            d_err[2].append(d_loss.data[0])
            d_acc[2].append(fake_score.data[0,0])
        #=============== Train the generator ===============#
        # Compute loss with fake images
        z = to_var(torch.randn(batch_size, 64))
        fake_images = G(z)
        outputs = D(0 * fake_images)
        
        # We train G to maximize log(D(G(z)) instead of minimizing log(1-D(G(z)))
        # For the reason, see the last paragraph of section 3. https://arxiv.org/abs/1406.2661
        g_loss = criterion(outputs, real_labels)
        
        # Backprop + Optimize
        D.zero_grad()
        G.zero_grad()
        g_loss.backward()
        g_optimizer.step()
        
        if (i+1) % 300 == 0:
            print('Epoch [%d/%d], Step[%d/%d], d_loss: %.4f, '
                  'g_loss: %.4f, D(x): %.2f, D(G(z)): %.2f' 
                  %(epoch, 5, i+1, 600, d_loss.data[0], g_loss.data[0],
                    real_score.data.mean(), fake_score.data.mean()))
    
    # Show a few real images
    if (epoch+1) == 1:
        images = images.view(images.size(0), 1, 28, 28)
        # save_image(denorm(images.data), './data/real_images.png')
        fig, axs = plt.subplots(1,3)
        plt.suptitle('real images')        
        for j in range(3):
            img = denorm(images[j].data)
            img = torch.cat([img, img, img], 0).float()
            show(img, axs[j])
    
    # Save sampled images
    fake_images = fake_images.view(fake_images.size(0), 1, 28, 28)
    #save_image(denorm(fake_images.data), './data/fake_images-%d.png' %(epoch+1))
    fig, axs = plt.subplots(1,3)
    plt.suptitle('fake images after epoch {}'.format(epoch+1))        
    for j in range(3):
        img = denorm(fake_images[j].data)
        img = torch.cat([img, img, img], 0).float()
        show(img, axs[j])
            
fig, ax = plt.subplots()
plt.ylabel('BCE')
plt.xlabel('training iterations')
plt.title('Discriminators error')
plt.plot(d_err[0], '-', c='r', label='After 1 epoch')
plt.plot(d_err[1], '--', c='b', label='After 3 epochs')
plt.plot(d_err[2], ':', c='g', label='After 5 epochs')
plt.legend()

fig, ax = plt.subplots()
plt.ylabel('Accuracy')
plt.xlabel('training iterations')
plt.title('Discriminators accuracy')
plt.plot(d_acc[0], '-', c='r', label='After 1 epoch')
plt.plot(d_acc[1], '--', c='b', label='After 3 epochs')
plt.plot(d_acc[2], ':', c='g', label='After 5 epochs')
plt.legend()
```
